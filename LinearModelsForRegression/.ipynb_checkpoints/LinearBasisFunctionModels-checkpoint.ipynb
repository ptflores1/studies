# Linear Basis Function Models
The simples linear model for regression is one that involves a linear combination of the input variables

$$y(\*x, \*w)=w_0+w_1x_1+\dots+w_Dx_D$$

where $\*x=(x_1, \dots, x_D)^T$. This is often simply known as *linear regression*. This model is a linear function of the parameters $w_0,\dots,w_D$. It is also a linear function on the input variables $x_i$, and this imposes significant limitations to the model. We therefore extend this models by considering linear combinations of fixed nonlinear functions $\phi_j$ of the input variables $x_i$, of the form

$$y(\*x, \*w) = w_0 + \sum_{j=1}^{M-1}w_j\phi_j(x)$$

where $\phi_j(\*x)$ are known as *basis functions*. Just to be clear, a basis function takes as input all the input variables $\*x=(x_i, \dots, x_D)$ and returns a transformation of them. $M-1$ is the amount of functions and is independent of the amount of input variables $D$. Therefore the number of parameters $\*w$ depend on the number of functions.

The parameter $w_0$ allows a fixed offset in the data and is sometimes called *bias*. It is often convenient to define a dummy basis function $\phi_0(\*x)=1$ so that we can write the linear combination as a dot product

$$y(\*x, \*w) = \sum_{j=0}^{M-1}w_j\phi_j(x) = \*w^T\*\phi(\*x)$$

where $\*w=(w_0,\dots,w_{M-1})^T$ and $\*\phi=(\phi_0,\dots,\phi_{M-1})^T$. Often times we apply some pre-processing or feature extraction to the original data variables. If the original variables comprise the vector $\*x$, then the features can be expressed in terms of the basis functions $\{\phi_j(\*x)\}$.

By using nonlinear basis functions we allow the function $y(\*x, \*w)$ to be a nonlinear function of the input variables. Functions of this form area called linear models, however, because this function is linear in $\*w$.

## Maximum likelihood and least squares

Let's assume we have a target variable $t$ that is given by a deterministic function $y(\*x, \*w)$ with additive Gaussian noise so that

$$t=y(\*x, \*w)+\epsilon$$

where $\epsilon$ is a zero mean Gaussian random variable with precision (inverse variance) $\beta$. Thus we can write

$$p(t|\*x,\*w,b)=\mathcal{N}(t|y(\*x,\*w),\beta^{-1})$$

If we assume a squared loss function 

$$E_D(\*w)=\frac{1}{2}\sum_{n-1}^N\{t_n-y(x_n, \*w)\}^2$$

then the optimal prediction for a new value of x, will be given by the conditional mean of the target variable. In the case of a Gaussian conditional distribution, the conditional mean will be simply

$$\mathbb{E}_D[t|\*x]=\int tp(t|\*x)dt=y(\*x,\*w)$$

Now consider a data set if inputs $\*X=\{\*x_1,\dots,\*x_N\}$ with corresponding target values $t_1,\dots,t_N$. We group the target variables $\{t_n\}$ into a column vector that we denote by $\*T$. Making the assumption that theses data points are drawn independently from the distribution, we obtain the following expression for the likelihood function, which is a function of the adjustable parameters $w$ and $\beta$, in the form

$$p(\*T|\*X,\*w,\beta)=\prod_{n=1}^N\mathcal{N}(t_n|\*w^T\phi(\*x_n), \beta^{-1})$$

In these models we are not modeling the distribution of the input variables $\*x$ so in order to maintain the notation uncluttered we drop them. Taking the logarithm of the likelihood function we have

$$ln\;p(\*T|\*w,\beta)=\sum_{n=1}^Nln\mathcal{N}(t_n|\*w^T\phi(\*x_n), \beta^{-1})$$
$$=\frac{N}{2}ln\;\beta-\frac{N}{2}ln(2\pi)-\beta E_D(\*w)$$

We can now use maximum likelihood to determine $\*w$ and $\beta$. Consider first the maximization with respect to $\*w$. The maximization of the likelihood function under a conditional Gaussian noise distribution for a linear model is equivalent to minimizing a sum-of-squares error function given by $E_D(\*w)$. The gradient of the log likelihood function takes the form

$$\nabla ln\;p(\*T|\*w,\beta)=\sum_{n=1}^N\{t_n-\*w^T\phi(\*x_n)\}\phi(\*x_n)^T$$

setting this gradient to zero gives

$$0=\sum_{n=1}^Nt_n\phi(\*x_n)^T -\*w^T\left(\sum_{n=1}^N\phi(\*x_n)\phi(\*x_n)^T\right)$$

solving for $w$ we obtain

$$\*w_{ML}=(\*\Phi^T\*\Phi)^{-1}\*\Phi^T\*T$$

which are known as the *normal equations* for the least squares problem. Here $\*\Phi$ is an $N\times M$ matrix, called the *design matrix*, whose elements are given by $\*\Phi_{nj} = \phi_j(\*x_n)$. This means that the i-th row corresponds to the i-th example evaluated in all the basis functions and the j-th column corresponds to the j-th basis function evaluated over all the examples.

We can also maximize the log likelihood function with respect to the noise precision parameter $\beta$, giving

$$\frac{1}{\beta_{ML}}=\frac{1}{N}\sum_{n=1}^N\{t_n-\*w_{ML}^T\phi(\*x_n)\}^2$$

and so we see that the inverse of the noise precision is given by the residual variance of the target values around the regression function.

### Example 1
For this example we consider some generated data that follows the line

$$y(x) = 3 x - 2$$

and assuming the data has some noise $\epsilon \sim \mathcal{N}(0, \beta^{-1})$ the distribution of our target variable $t$ is

$$t \sim \mathcal{N}(y(x), \beta^{-1})$$

When generating the data we will use $\beta^{-1} = 2$.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
noise = np.random.normal(0, 2, 100)
y = 3 * x - 2 + noise

plt.figure(dpi=150)
plt.scatter(x, y, label="$3x-2+noise$")
plt.plot(x, 3*x-2, label="True function", c="r")
plt.title("Random Linear Data")
plt.legend()
plt.tight_layout()
plt.show()
```

To fit the data we will use the simplest basis function for $x$ which is the identtity $\phi_1(x)=x$, and using the dummy function we get 

$$\*\Phi(x) = 
\begin{pmatrix}
1 & x_1 \\
1 & x_2 \\ 
\vdots & \vdots \\ 
1 & x_N \\ 
\end{pmatrix}$$

Using this basis function means that we assume our data can be represented by a function $\hat y(x)=w_0+w_1x$, which is exactly the case.
Just for the sake of sanity, let's check if the dimensions of the estimation of the weights $w_{ML}$ check out

$$\*w_{ML}=(\*\Phi^T\*\Phi)^{-1}\*\Phi^T\*T$$
$$\*w_{ML}=((2\times 100)\times(100\times 2))(2\times 100)\times(100\times 1) = (2\times 1)$$

which is completely fine given that we want two weights $w_0, w_1$. Let's compute both of this weights

```python
design_matrix = np.stack([np.ones(100), x], axis=1)
design_matrix.shape
```

The @ symbol in numpy is used to do matrix multiplication.

```python
estimated_weights = np.linalg.inv(design_matrix.T @ design_matrix) @ design_matrix.T @ y.reshape(-1 ,1)
estimated_weights
```

Now having $\*w_{ML}$ we can compute $\beta_{ML}$. We do all the $\*w_{ML}^T\phi(\*x_n)$ multiplications as a matrix multiplication, since $\phi(\*x_n)$ is a column of $\*\Phi$ we need to transpose it.

```python
estimated_beta_inv = ((y.reshape(-1 ,1).T - estimated_weights.T @ design_matrix.T)**2).mean()
estimated_beta_inv
```

Let's make a plot to visualize how our predictions look compared with the true function.

```python
c = plt.rcParams['axes.prop_cycle'].by_key()['color']

plt.figure(dpi=150)
plt.scatter(x, y, label="$3x-2+noise$", alpha=0.5, linewidth=0)
plt.plot(x, 3*x-2, label="True function", c="r")

# Plot estimated function with the weights learned
w0, w1 = estimated_weights
plt.plot(x, w0 + w1*x, label="Estimated function", c=c[2])
plt.fill_between(
            x, w0 + w1*x - estimated_beta_inv, w0 + w1*x + estimated_beta_inv,
            alpha=0.3,
            edgecolor=None,
            facecolor=c[2],
            linewidth=0,
            zorder=1,
            label="1 std")

plt.title("Results")
plt.legend()
plt.tight_layout()
plt.show()
```

### Example 2
Now let's explore a more difficult function to learn. For this example we will be using the sine function as the ground truth

$$y(x) = 5sin(x)+1$$

and again we assume some random noise $\epsilon \sim \mathcal{N}(0, \beta^{-1})$ and with that the distribution of the target variable is

$$t\sim \mathcal{N}(y(x),\beta^{-1})$$

```python
x = np.linspace(0, 10, 100)
noise = np.random.normal(0, 1, 100)
y = 5*np.sin(x) + 1 + noise

plt.figure(dpi=150)
plt.scatter(x, y, label="$5 sin(x) + 1 + noise$")
plt.plot(x, 5*np.sin(x)+1, label="True function", c="r")
plt.title("Random Sinusoidal Data")
plt.legend()
plt.tight_layout()
plt.show()
```

To fit this data we will use some different basis functions, but let's start by using the same one that in the previous example

$$\*\Phi(x) = 
\begin{pmatrix}
1 & x_1 \\
1 & x_2 \\ 
\vdots & \vdots \\ 
1 & x_N \\ 
\end{pmatrix}$$

```python
design_matrix = np.stack([np.ones(100), x], axis=1)
estimated_weights = np.linalg.inv(design_matrix.T @ design_matrix) @ design_matrix.T @ y.reshape(-1 ,1)
estimated_beta_inv = ((y.reshape(-1 ,1).T - estimated_weights.T @ design_matrix.T)**2).mean()
```

```python
plt.figure(dpi=150)
plt.scatter(x, y, label="$5 sin(x) + 1 + noise$", alpha=0.5, linewidth=0)
plt.plot(x, 5*np.sin(x)+1, label="True function", c="r")

# Plot estimated function with the weights learned
w0, w1 = estimated_weights
plt.plot(x, w0 + w1*x, label="Estimated function", c="tab:green")
plt.fill_between(
            x, w0 + w1*x - estimated_beta_inv, w0 + w1*x + estimated_beta_inv,
            alpha=0.3,
            edgecolor=None,
            facecolor="tab:green",
            linewidth=0,
            zorder=1,
            label="1 std")

plt.title("Results")
plt.legend()
plt.tight_layout()
plt.show()
```

Obviously the basis function used is way too simple and can not fit the data, we need to use more complex basis functions. We now explore basis functions in the form of powers of the input variable, so we can use $x^2, x^3,\dots$. Let's define the following design matrices

$$\*\Phi_1(x) = 
\begin{pmatrix}
1 & x_1 & x_1^2\\
1 & x_2 & x_2^2\\ 
\vdots & \vdots& \vdots \\ 
1 & x_N & x_N^2\\ 
\end{pmatrix}
\quad
\*\Phi_2(x) = 
\begin{pmatrix}
1 & x_1 & x_1^2& x_1^3\\
1 & x_2 & x_2^2& x_2^3\\ 
\vdots & \vdots& \vdots& \vdots \\ 
1 & x_N & x_N^2& x_N^3\\ 
\end{pmatrix}
\quad
\*\Phi_3(x) = 
\begin{pmatrix}
1 & x_1 & x_1^2& x_1^3& x_1^4\\
1 & x_2 & x_2^2& x_2^3& x_2^4\\ 
\vdots & \vdots& \vdots& \vdots& \vdots \\ 
1 & x_N & x_N^2& x_N^3& x_N^4\\ 
\end{pmatrix}
$$

that means that we will try to fit 3 different functions
$$\hat y_1(x)=w_0 + w_1x+w_2x^2$$
$$\hat y_2(x)=w_0 + w_1x+w_2x^2+w_3x^3$$
$$\hat y_3(x)=w_0 + w_1x+w_2x^2+w_3x^3+w_4x^4$$

each one correspondig to the design matrices.

```python
design_matrix_1 = np.stack([np.ones(100), x, x**2], axis=1)
estimated_weights_1 = np.linalg.inv(design_matrix_1.T @ design_matrix_1) @ design_matrix_1.T @ y.reshape(-1 ,1)
estimated_beta_inv_1 = ((y.reshape(-1 ,1).T - estimated_weights_1.T @ design_matrix_1.T)**2).mean()
```

```python
design_matrix_2 = np.stack([np.ones(100), x, x**2, x**3], axis=1)
estimated_weights_2 = np.linalg.inv(design_matrix_2.T @ design_matrix_2) @ design_matrix_2.T @ y.reshape(-1 ,1)
estimated_beta_inv_2 = ((y.reshape(-1 ,1).T - estimated_weights_2.T @ design_matrix_2.T)**2).mean()
```

```python
design_matrix_3 = np.stack([np.ones(100), x, x**2, x**3, x**4], axis=1)
estimated_weights_3 = np.linalg.inv(design_matrix_3.T @ design_matrix_3) @ design_matrix_3.T @ y.reshape(-1 ,1)
estimated_beta_inv_3 = ((y.reshape(-1 ,1).T - estimated_weights_3.T @ design_matrix_3.T)**2).mean()
```

```python
fig, axes = plt.subplots(1, 3, sharex=True, sharey=True, dpi=100, figsize=(12, 4))

for i, ax in enumerate(axes):
    ax.scatter(x, y, label="$5 sin(x) + 1 + noise$", alpha=0.5, linewidth=0)
    ax.plot(x, 5*np.sin(x)+1, label="True function", c="r")

# Plot estimated function with the weights learned
w0, w1, w2 = estimated_weights_1
prediction = w0 + w1*x + w2*x**2
axes[0].plot(x, prediction, label="Estimated function", c="tab:green")
axes[0].fill_between(
            x, prediction - estimated_beta_inv_1, prediction + estimated_beta_inv_1,
            alpha=0.3,
            edgecolor=None,
            facecolor="tab:green",
            linewidth=0,
            zorder=1,
            label="1 std")
axes[0].legend()
axes[0].set_title("$w_0 + w_1x + w_2 x^2$")

w0, w1, w2, w3 = estimated_weights_2
prediction = w0 + w1*x + w2*x**2 + w3*x**3
axes[1].plot(x, prediction, label="Estimated function", c="tab:green")
axes[1].fill_between(
            x, prediction - estimated_beta_inv_2, prediction + estimated_beta_inv_2,
            alpha=0.3,
            edgecolor=None,
            facecolor="tab:green",
            linewidth=0,
            zorder=1,
            label="1 std")
axes[1].legend()
axes[1].set_title("$w_0 + w_1x + w_2 x^2 +w_3x^3$")

w0, w1, w2, w3, w4 = estimated_weights_3
prediction = w0 + w1*x + w2*x**2 + w3*x**3 + w4*x**4
axes[2].plot(x, prediction, label="Estimated function", c="tab:green")
axes[2].fill_between(
            x, prediction - estimated_beta_inv_3, prediction + estimated_beta_inv_3,
            alpha=0.3,
            edgecolor=None,
            facecolor="tab:green",
            linewidth=0,
            zorder=1,
            label="1 std")
axes[2].legend()
axes[2].set_title("$w_0 + w_1x + w_2 x^2 +w_3x^3+w_4x^4$")

plt.tight_layout()
plt.show()
```

We can clearly see that by increasing the model complexity the predictions improve, the best one so far is the model that transforms $x$ up to $x^2$. We know that the data is generated from the function $y(x) = 5sin(x)+1$, in an actual real life scenario we would not know this function but let's use as basis function $sin(x)$ anyway and see how much the model improves.

We want to fit a function of the form

$$\hat y(x) = w_0+sin(x)+1$$

the corresponding desing matrix would be

$$\*\Phi(x) = 
\begin{pmatrix}
1 & sin(x_1) \\
1 & sin(x_2) \\ 
\vdots & \vdots \\ 
1 & sin(x_N) \\ 
\end{pmatrix}$$

```python
design_matrix_sin = np.stack([np.ones(100), np.sin(x)], axis=1)
estimated_weights_sin = np.linalg.inv(design_matrix_sin.T @ design_matrix_sin) @ design_matrix_sin.T @ y.reshape(-1 ,1)
estimated_beta_inv_sin = ((y.reshape(-1 ,1).T - estimated_weights_sin.T @ design_matrix_sin.T)**2).mean()
```

```python
plt.figure(dpi=150)
plt.scatter(x, y, label="$5 sin(x) + 1 + noise$", alpha=0.5, linewidth=0)
plt.plot(x, 5*np.sin(x)+1, label="True function", c="r")

# Plot estimated function with the weights learned
w0, w1 = estimated_weights_sin
prediction = w0 + w1*np.sin(x)
plt.plot(x, prediction, label="Estimated function", c="tab:green")
plt.fill_between(
            x, prediction - estimated_beta_inv_sin, prediction + estimated_beta_inv_sin,
            alpha=0.3,
            edgecolor=None,
            facecolor="tab:green",
            linewidth=0,
            zorder=1,
            label="1 std")

plt.title("Results")
plt.legend()
plt.tight_layout()
plt.show()
```

It is evident that this model is way better than the degree 4 polynomial. We can check how close the weights got to the real ones, we expect to get $\hat w_0=1$ and $\hat w_1=5$

## Sequential Learning
Batch techniques, such as the maximum likelihood solution, which involve processing the entire training set in one go, can be computationally costly for large data sets. To avoid this problem we can use *sequential* algorithms, also known as  *on-line* algorithms, in which the data points are cosidered one at a time, and the model parameters updated after each such presetnation.

We can obtain a sequential learning algorithm by applying the technique of *stochastic gradient descent*, also known as *sequential gradient descent*, as follows. If the error function comprises a sum over data points $E=\sum_n E_n$, then after presetation fo pattern $n$, the stochastic gradient descent algorithm updates the parameter vecotr $\*w$ using

$$\*w^{(\tau+1)}=\*w^{(\tau)} - \eta \nabla E_n$$

where $\tau$ denotes the itration number, and $\eta$ is a learning rate parameter. The value of $\*w$ is initialized to some starting vector $\*w^0$. For the case of the sum-of-squares error function, this gives 

$$\*w^{(\tau+1)}=\*w^{(\tau)} - \eta (t_n - \*w^{(\tau)T}\phi_n)\phi_n$$

where $\phi_n = \phi(x_n)$. This is known as *least-mean-squares* or the *LMS algorithm*. The learning rate parameter $\eta$ controls how fast the algorithm learns, but has to be chosen carefully because too large of a value may cause the algorithm to not learn at all.
